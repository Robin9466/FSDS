````python
# 1. Simple Number Generator

# Create a generator function called simple_numbers that yields numbers from 1 to 10. Use the generator to print each number.
def simple_numbers():
    for i in range(1, 11):
        yield i

# Create an instance of the generator and use a loop to get the values
gen = simple_numbers()

for num in gen:
    print(num)
````

````python
# 2. Odd Number Generator

# Write a generator called odd_numbers that yields odd numbers between 1 and 20. Iterate through the generator and print the values.
def odd_numbers():
  for i in range(1, 21):
    if i % 2 != 0:
        yield i

# Create an instance of generator and use a loop to get the values
gen = odd_numbers()

for num in gen:
    print(num)
````

````python
# 3. Fibonacci Sequence Generator

# Create a generator function fibonacci_generator that yields the Fibonacci sequence up to a certain number of terms (e.g., first 10 terms). Test it by printing the sequence.
def fibonacci_generator(n):
  a, b = 0, 1

  for _ in range(n):
    yield a

    a, b = b, a+b

# Generate the first 10 fibanocci series
for num in fibonacci_generator(10):
    print(num)
````

````python
# 4. File Line Reader (Simulating Large Files)

# Write a generator function file_reader that simulates reading a large file by yielding one line at a time from a list of lines (you can use a list of strings as a mock file). Print each line using the generator.
def file_reader(lines):
  for line in lines:
    yield line

# Mock file content (list of strings)
mock_file = [
    "Line 1: This is the first line.",
    "Line 2: This is the second line.",
    "Line 3: This is the third line.",
    "Line 4: This is the fourth line."
]

# Create an instance of the generator
gen = file_reader(mock_file)

# Printing the mock file
for line in gen:
    print(line)
````

````python
# 5. Infinite Even Number Generator

# Create an infinite generator even_numbers that yields even numbers starting from 2. Use a loop to print only the first 10 even numbers.
def even_numbers():
    i = 2
    while True:
        yield i
        i += 2

# Creating an instance of the generator:
gen = even_numbers()

# Printing the first 10 even numbers
for _ in range(10):
    print(next(gen))
````

````python
# 6. Reverse List Generator

# Implement a generator function reverse_list that takes a list and yields its elements in reverse order. Test it by passing a list and printing the elements one by one.

def reverse_list(lst):
    # Loop through the list in reverse order and yield each element
    for i in range(len(lst)-1, -1, -1):
        yield lst[i]

my_list = ['my ','name ', 'is ', 'Robin']

# Create an instane of the generator
gen = reverse_list(my_list)

# Printing the reversed elements from the list one by one:
for element in gen:
    print(element)
````

````python
# # 7. Prime Number Generator
# Write a generator prime_numbers that yields prime numbers indefinitely. Use the generator to print the first 10 prime numbers.

def is_prime(num):
    if num <= 1:
        return False
    if num in (2,3):
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i+2) == 0:
            return False
        i+= 6
    return True

def prime_numbers():
    num =2
    while True:
        if is_prime(num):
            yield num
        num +=1

prime_gen = prime_numbers()
for _ in range(10):
    print(next(prime_gen))
````
