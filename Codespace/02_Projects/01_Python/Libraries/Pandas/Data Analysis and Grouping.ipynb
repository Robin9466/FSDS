1. Group Data Based on a Column
Task: Group data by a specific column (e.g., group people by country) and perform aggregations like calculating the average age.

```python
import pandas as pd

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Country': ['USA', 'Canada', 'USA', 'Canada', 'UK', 'USA'],
    'Age': [25, 30, 35, 40, 28, 22]
}

df = pd.DataFrame(data)
df
```

Group people by the Country and calculate the average Age.

```python
# Solution
average_age_by_country = df.groupby('Country')['Age'].mean()
print(average_age_by_country)
```

```python
# Sample dataset
import pandas as pd

data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Sales': [100, 200, 150, 300, 250, 180]
}

df_product_sales = pd.DataFrame(data)  # Create a new DataFrame for product sales

# Calculate total sales by product using the new DataFrame
total_sales_by_product = df_product_sales.groupby('Product')['Sales'].sum()
print(total_sales_by_product)
```

3. Create a Pivot Table
Task: Construct a pivot table from a simple dataset.

```python
# Sample dataset
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03', '2024-01-03'],
    'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles', 'New York', 'Los Angeles'],
    'Sales': [250, 300, 200, 450, 500, 350]
}

df = pd.DataFrame(data)
df
```

```python
# Solution
pivot_table = df.pivot_table(values='Sales', index='Date', columns='City', aggfunc='sum')
print(pivot_table)
```

4. Perform Statistical Analysis
Task: Compute basic statistical measures such as mean, median, mode, and standard deviation.

```python
# Sample dataset
data = {
    'Category': ['Electronics', 'Groceries', 'Electronics', 'Groceries', 'Clothing', 'Clothing'],
    'Price': [1200, 50, 1500, 60, 200, 250]
}

df = pd.DataFrame(data)
df
```

Exercise
Calculate:
Mean of prices.
Median of prices.
Mode of categories.
Standard deviation of prices.

```python
# Solutions
mean_price = df['Price'].mean()
median_price = df['Price'].median()
mode_category = df['Category'].mode()[0]
std_dev_price = df['Price'].std()

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Category: {mode_category}")
print(f"Standard Deviation of Prices: {std_dev_price}")
```
