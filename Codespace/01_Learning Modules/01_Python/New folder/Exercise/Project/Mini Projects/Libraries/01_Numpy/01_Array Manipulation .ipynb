```python
'''
Mini-Project: Create a data merger that stacks student scores (row-wise) from two different tests and then splits them by groups of students.
'''
import numpy as np

# Step 1: Creating mock data using Numpy arrays
# Data for Test 1 (5 students, 2 subjects)
data_test1 = np.array([
    [85, 90],  # S1 - [Math Score, English Score]
    [78, 82],  # S2
    [92, 89],  # S3
    [88, 85],  # S4
    [76, 80]   # S5
])

data_test2 = np.array([
    [91, 88],  # S6
    [74, 79],  # S7
    [89, 92],  # S8
    [95, 87],  # S9
    [80, 81]   # S10
])

# Step 2: Merge Data
merged_data = np.vstack((data_test1, data_test2))

# Step 3: Splitting
group_size = 5
student_groups = np.array_split(merged_data, len(merged_data)//group_size)
print(type(student_groups))
# Display each group using enumerate
for index, group in enumerate(student_groups):
    print(f"Group {index + 1}:")
    print(group)
    print("\n")
```
