#### **Exception Handling Assignment:**
1. Write a Python script that asks the user for two numbers and performs division.
2. Handle the following exceptions:
   - `ValueError` if the user enters non-numeric input.
   - `ZeroDivisionError` if the user tries to divide by zero.
3. Ensure that appropriate messages are displayed for each type of exception.

Feel free to reach out if you need any clarifications or further help with the assignments!

```python
def main():
    try:
        # Get user input and perform division
        num1, num2 = map(float, input("Enter two numbers separated by space: ").split())
        result = num1 / num2
        print(f"The result of {num1} divided by {num2} is {result:.2f}")

    except (ValueError, ZeroDivisionError) as e:
        if isinstance(e, ValueError):
            print("Error: Please enter valid numbers separated by spaces.")
        elif isinstance(e, ZeroDivisionError):
            print("Error: Division by zero is not allowed.")

# Call the main function
main()
```
