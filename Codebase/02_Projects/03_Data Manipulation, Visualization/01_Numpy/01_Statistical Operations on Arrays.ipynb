```python
'''
Mini-Project: Generate a dataset of exam scores, calculate the mean, median, and standard deviation, then visualize the data distribution using a histogram.
'''

import numpy as np
import matplotlib.pyplot as plt

# exam_scores = np.array([78, 99, 97, 69, 53, 78, 98, 97, 67, 97])

exam_scores = np.random.randint(80, 101, size=100)

# Calculate statistical measures
mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)
std_deviation = np.std(exam_scores)

# Compute histogram using numpy.histogram
hist, bin_edges = np.histogram(exam_scores, bins=10)

# Plot the histogram
plt.figure(figsize=(8,6))
plt.hist(exam_scores, bins=10, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Histogram of exam scores')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_score:.2f}')
plt.axvline(median_score, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median_score:.2f}')
plt.legend()
plt.grid(True)
plt.show()

# Output the statistical results
mean_score, median_score, std_deviation
```
