Exercise 1: Create a List of Squares

Generate a list of squares for numbers from 1 to 10.

```python
list_of_squares = [x**2 for x in range(10)]
print(list_of_squares)
```

Exercise 2: Filter Even Numbers

Create a list of even numbers from 1 to 20.

```python
[x for x in range(1,21) if x%2==0]
```

Exercise 3: Convert Strings to Uppercase

Given a list of words, create a new list with each word converted to uppercase.

```python
def uppercase(word):
    return word.capitalize()

words = ['ram', 'shyam', 'rohan', 'geeta', 'meena']
[uppercase(word) for word in words]
```
