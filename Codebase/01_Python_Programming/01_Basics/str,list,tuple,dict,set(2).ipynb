```python
l = [3, 4, 5, 6, 7, [23, 456, 67, 8, 78, 78], [345, 56, 87, 8, 98, 9], (234, 6657, 6), {"key1": "sudh", 234: [23, 45, 656]}]

all_234_instances = []

for item in l:
    if item == 234:
        all_234_instances.append(item)
    elif isinstance(item, list) or isinstance(item, tuple):
        for i in item:
            if i == 234:
                all_234_instances.append(i)
    elif isinstance(item, dict):
        if 234 in item:
            all_234_instances.append(234)
        for value in item.values():
            if isinstance(value, (list, tuple)):
                for v in value:
                    if v == 234:
                        all_234_instances.append(v)
            else:
                if value == 234:
                    all_234_instances.append(value)

print(all_234_instances)
```
