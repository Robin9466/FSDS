```python
import sys
while True:
    try:
        a=int(input("first number"))
        b=int(input("second number"))
        c=a/b
        print(f"div: {c}")
        break
    except:
        a,b,c=sys.exc_info()
        print(f"exception class: {a}")
        print("exception message",b)
        print("line number",c.tb_lineno)
```
