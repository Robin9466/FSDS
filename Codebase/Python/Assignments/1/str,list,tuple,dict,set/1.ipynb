```python
'''
l = [3,4,5,6,7 , [23,456,67,8,78,78] , [345,56,87,8,98,9] , (234,6657,6) , {"key1" :"sudh" , 234:[23,45,656]}]

2. try to access 234 out of this list
'''
l = [3,4,5,6,7 , [23,456,67,8,78,78] , [345,56,87,8,98,9] , (234,6657,6) , {"key1" :"sudh" , 234:[23,45,656]}]

access_234_elements = []
for i in l:
    if isinstance(i, int) and i == 234:
        access_234_elements.append(i)
    if isinstance(i, (list,tuple)):
        for item in i:
            if item == 234:
                access_234_elements.append(item)
    if isinstance(i, dict):
        for item in i.keys():
            if item == 234:
                access_234_elements.append(item)
print(access_234_elements)
```
