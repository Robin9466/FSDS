```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([2, 4, 6, 8, 10])
Y_positive = np.array([30, 28, 26, 24, 22])  # Perfect positive correlation

plt.scatter(X, Y_positive, color='blue')
plt.title('Perfect Positive Correlation (\u03C1 = 1)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
Y_negative = np.flip(Y_positive)  # Perfect negative correlation

plt.scatter(X, Y_negative, color='red')
plt.title('Perfect Negative Correlation (\u03C1 = -1)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
Y_slight_negative = np.array([30, 28, 26, 24, 23])  # Slight negative correlation

plt.scatter(X, Y_slight_negative, color='green')
plt.title('Slight Negative Correlation (\u03C1 = -0.2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
