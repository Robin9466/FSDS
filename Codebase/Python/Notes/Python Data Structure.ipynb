# **Python Data Structures: Comprehensive Overview**

---

## **1. Lists**

### **1.1. Definition**
- Ordered, mutable collections of items.
- Can hold mixed data types.

### **1.2. Syntax**
- **Creating a List**:
  ```python
  my_list = [1, 2, 3, "apple", "banana"]
  ```

### **1.3. Key Operations**
- **Accessing Elements**:
  ```python
  print(my_list[0])  # Output: 1
  ```
- **Modifying Elements**:
  ```python
  my_list[2] = "orange"
  ```
- **Appending Elements**:
  ```python
  my_list.append("grape")
  ```
- **Removing Elements**:
  ```python
  my_list.remove("apple")
  ```
- **Slicing**:
  ```python
  sub_list = my_list[1:4]
  ```

---

## **2. Sets**

### **2.1. Definition**
- Unordered collections of unique items.
- Do not allow duplicate elements.

### **2.2. Syntax**
- **Creating a Set**:
  ```python
  my_set = {1, 2, 3, "apple", "banana"}
  ```

### **2.3. Key Operations**
- **Adding Elements**:
  ```python
  my_set.add("grape")
  ```
- **Removing Elements**:
  ```python
  my_set.remove("apple")
  ```
- **Set Operations**:
  - **Union**:
    ```python
    another_set = {3, 4, 5}
    union_set = my_set | another_set
    ```
  - **Intersection**:
    ```python
    intersection_set = my_set & another_set
    ```
  - **Difference**:
    ```python
    difference_set = my_set - another_set
    ```

---

## **3. Dictionaries**

### **3.1. Definition**
- Unordered collections of key-value pairs.
- Keys are unique and immutable; values can be any type.

### **3.2. Syntax**
- **Creating a Dictionary**:
  ```python
  my_dict = {"name": "Alice", "age": 25}
  ```

### **3.3. Key Operations**
- **Accessing Values**:
  ```python
  print(my_dict["name"])  # Output: Alice
  ```
- **Modifying Values**:
  ```python
  my_dict["age"] = 26
  ```
- **Adding Key-Value Pairs**:
  ```python
  my_dict["city"] = "New York"
  ```
- **Removing Key-Value Pairs**:
  ```python
  del my_dict["city"]
  ```
- **Keys and Values**:
  ```python
  keys = my_dict.keys()   # Output: dict_keys(['name', 'age'])
  values = my_dict.values()  # Output: dict_values(['Alice', 26])
  ```

---

## **4. Strings**

### **4.1. Definition**
- Sequences of characters enclosed in quotes.
- Immutable.

### **4.2. Syntax**
- **Creating a String**:
  ```python
  my_string = "Hello, World!"
  ```

### **4.3. Key Operations**
- **Accessing Characters**:
  ```python
  print(my_string[0])  # Output: H
  ```
- **Slicing**:
  ```python
  sub_string = my_string[0:5]  # Output: Hello
  ```
- **Concatenation**:
  ```python
  full_string = my_string + " How are you?"
  ```
- **Repetition**:
  ```python
  repeated_string = my_string * 2  # Output: Hello, World!Hello, World!
  ```
- **Methods**:
  - **`upper()`**:
    ```python
    print(my_string.upper())  # Output: HELLO, WORLD!
    ```
  - **`lower()`**:
    ```python
    print(my_string.lower())  # Output: hello, world!
    ```
  - **`split()`**:
    ```python
    words = my_string.split(", ")  # Output: ['Hello', 'World!']
    ```
  - **`replace()`**:
    ```python
    new_string = my_string.replace("World", "Python")  # Output: Hello, Python!
    ```

## **5. Tuples**

### **5.1. Definition**
- Ordered, immutable collections of items.
- Can contain mixed data types.

### **5.2. Syntax**
- **Creating a Tuple**:
  ```python
  my_tuple = (1, 2, 3, "apple", "banana")
  ```

### **5.3. Key Operations**
- **Accessing Elements**:
  ```python
  print(my_tuple[1])  # Output: 2
  ```

## **6. Frozensets**

### **6.1. Definition**
- Immutable sets.
- Unordered collections of unique items.

### **6.2. Syntax**
- **Creating a Frozenset**:
  ```python
  my_frozenset = frozenset([1, 2, 3, 4])
  ```

### **6.3. Key Operations**
- **Set Operations**:
  ```python
  another_frozenset = frozenset([3, 4, 5, 6])
  union_set = my_frozenset | another_frozenset
  intersection_set = my_frozenset & another_frozenset
  difference_set = my_frozenset - another_frozenset
  ```
- **Immutability**:
  ```python
  # Attempting to modify a frozenset will raise an error
  # my_frozenset.add(7)  # Raises AttributeError
  ```

## **7. Arrays**

### **7.1. Definition**
- Similar to lists but more memory-efficient.
- Intended for numerical data of a single type.

### **7.2. Syntax**
- **Creating an Array**:
  ```python
  import array
  my_array = array.array('i', [1, 2, 3, 4])  # 'i' denotes integer type
  ```

### **7.3. Key Operations**
- **Accessing Elements**:
  ```python
  print(my_array[1])  # Output: 2
  ```
- **Appending Elements**:
  ```python
  my_array.append(5)
  ```
- **Removing Elements**:
  ```python
  my_array.remove(3)
  ```

---

## **8. Deques**

### **8.1. Definition**
- Double-ended queues that allow appends and pops from both ends.
- Part of the `collections` module.

### **8.2. Syntax**
- **Creating a Deque**:
  ```python
  from collections import deque
  my_deque = deque([1, 2, 3])
  ```

### **8.3. Key Operations**
- **Appending to Ends**:
  ```python
  my_deque.append(4)       # Appends to the right
  my_deque.appendleft(0)   # Appends to the left
  ```
- **Popping from Ends**:
  ```python
  my_deque.pop()           # Removes from the right
  my_deque.popleft()       # Removes from the left
  ```

---

## **9. Named Tuples**

### **9.1. Definition**
- Tuples with named fields for better readability.
- Part of the `collections` module.

### **9.2. Syntax**
- **Creating a Named Tuple**:
  ```python
  from collections import namedtuple
  Person = namedtuple('Person', ['name', 'age'])
  p = Person(name="Alice", age=30)
  ```

### **9.3. Key Operations**
- **Accessing Fields**:
  ```python
  print(p.name)  # Output: Alice
  print(p.age)   # Output: 30
  ```
- **Converting to a Dictionary**:
  ```python
  p_dict = p._asdict()
  ```
