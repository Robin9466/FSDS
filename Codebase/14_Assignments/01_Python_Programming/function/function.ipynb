````python
#  q1 : Try to print a prime number in between 1 to 1000
import math

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for c in range(2, math.isqrt(n) + 1):  # Check divisibility up to sqrt(n)
        if n % c == 0:
            return False
    return True

# Loop through numbers from 1 to 1000 and print prime numbers
for num in range(1, 1001):
    if is_prime(num):
        print(num, end = ' ')
````

````python
#  q2 : Try to write a function which  is equivalent  to print function in python

import sys

def my_print(*args, sep = ' ', end = '\n', file = None): # Added file = None as a parameter
    output = sep.join(map(str, args))

    # If no file is specified, default to sys.stdout (the screen)
    if file is None:
        file = sys.stdout

    # Write the output to the file (or screen)
    file.write(output + end)

# Example usage:
my_print("Hello", "world!")  # Output: Hello world!
my_print(1, 2, 3, sep='-', end=' END\n')
````

Let’s break down `sep.join(map(str, args))` step by step.

### 1. `map(str, args)`
- **`args`**: This is a tuple of arguments that can be any data types (e.g., strings, integers, floats, etc.). When you pass multiple arguments to the function, they are collected in `*args`.
  Example: `args = (1, "world", 3.14)`
  
- **`map(str, args)`**: The `map()` function applies the `str` function to each element of `args`. This converts each element into a string, since the `print()` function can only print strings.
  
  Example:
  ```python
  map(str, args)  # Converts 1 to "1", "world" stays the same, 3.14 becomes "3.14"
  ```

  The result of `map()` is an iterable of string representations of the `args` elements:
  ```python
  ["1", "world", "3.14"]
  ```

### 2. `sep.join()`
- **`sep`**: This is the separator provided by the user (default is a space `' '`), which is used to join the strings together.
  
- **`sep.join(iterable)`**: The `join()` method takes an iterable (in this case, the list of strings produced by `map()`) and concatenates the strings with `sep` in between each element.

  Example:
  ```python
  sep = " "
  sep.join(["1", "world", "3.14"])
  # Result: "1 world 3.14"
  ```

  If `sep` were something else, like a hyphen (`'-'`):
  ```python
  sep = "-"
  sep.join(["1", "world", "3.14"])
  # Result: "1-world-3.14"
  ```

### Overall Example:
If you call:
```python
my_print(1, "world", 3.14, sep="-")
```
The sequence of operations is:
1. `map(str, args)` turns `args` into `["1", "world", "3.14"]`.
2. `sep.join(...)` joins these strings with `"-"` as the separator, resulting in `"1-world-3.14"`.

### Summary:
- `map(str, args)` converts all elements of `args` to strings.
- `sep.join(...)` joins these string elements using the `sep` separator, resulting in a final single string that can be printed or written to a file.

````python
#  q3 : Try to write a function which is a replica of list append , extend and pop function

# replica of append:

def my_append(lst, item):
    lst[len(lst):] = [item] # Add the item at the end of the list

my_list = [1, 2, 3, 4]
my_append(my_list, 5)
print(my_list)


#  Replica of extend:
def my_extend(lst, iterable):
    lst[len(lst):] = iterable

my_extend(my_list, [6, 7, 8, 9, 0])
print(my_list)


# Replica of pop:

def my_pop(lst, index = None):
    if index is None:
        index = len(lst) - 1
    item = lst[index]  # Get the item from the list, which is to be popped up
    del lst[index]

    return item

popped_item = my_pop(my_list)
print(f"{popped_item}")
print(my_list)
````

To replicate the behavior of the `append()`, `extend()`, and `pop()` methods for Python lists, we can write custom functions. Below are the steps and explanations for each function.

### 1. **`append()`**:
The `append()` function adds a single element to the end of the list.

#### Replica of `append()`

```python
def my_append(lst, item):
    lst[len(lst):] = [item]  # Add the item at the end of the list
```

- **Explanation**:
  - `lst[len(lst):] = [item]`: This slices the list starting from its end (`len(lst):`) and replaces that part with a new list containing `item`, effectively appending it to the end.
  
#### Example:

```python
my_list = [1, 2, 3]
my_append(my_list, 4)
print(my_list)  # Output: [1, 2, 3, 4]
```

---

### 2. **`extend()`**:
The `extend()` function adds all elements from an iterable (like another list) to the end of the current list.

#### Replica of `extend()`

```python
def my_extend(lst, iterable):
    lst[len(lst):] = iterable  # Add all elements from the iterable to the end
```

- **Explanation**:
  - This uses the same slicing method but replaces the empty slice at the end of `lst` with the entire iterable (list, set, etc.).
  
#### Example:

```python
my_list = [1, 2, 3]
my_extend(my_list, [4, 5, 6])
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]
```

---

### 3. **`pop()`**:
The `pop()` function removes and returns the last element from the list. Optionally, it can remove the element at a specified index.

#### Replica of `pop()`

```python
def my_pop(lst, index=None):
    if index is None:
        index = len(lst) - 1  # Default is to pop the last element
    item = lst[index]  # Get the element at the specified index
    del lst[index]     # Remove the element from the list
    return item        # Return the popped element
```

- **Explanation**:
  - If `index` is not specified, the function defaults to removing the last element (`index = len(lst) - 1`).
  - It retrieves the element at `index`, deletes it from the list using `del lst[index]`, and then returns the element.
  
#### Example:

```python
my_list = [1, 2, 3, 4]
popped_item = my_pop(my_list)  # Pops the last element
print(my_list)  # Output: [1, 2, 3]
print(popped_item)  # Output: 4
```

---

### Summary of Functions:
- **`my_append(lst, item)`**: Adds a single item to the end of the list.
- **`my_extend(lst, iterable)`**: Adds all elements from an iterable to the end of the list.
- **`my_pop(lst, index=None)`**: Removes and returns the element at `index` (defaults to the last element if no index is provided).

These functions behave like their built-in counterparts (`append()`, `extend()`, and `pop()`), but using custom implementations.

````python
# 1. Basic Arithmetic Operation

addition = lambda x,y: x+y

result = addition(5,3)
print(result)
````

````python
# 2. Sorting a List of Tuples by Second Element

# Given a list of tuples, sort the list by the second element in each tuple. Use a lambda function as the key for sorting.


data = [(1, 5), (3, 2), (4, 4), (2, 3)]

sorted_data = sorted(data, key = lambda x: x[1])
print(sorted_data)

````

````python
# 3. Filtering Even Numbers from a List

# Use a lambda function to filter out even numbers from a list of integers.

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
````

````python
# 4. Finding the Maximum in a List
# Write a lambda function to find the maximum value in a list of numbers.

from functools import reduce

values = [10, 20, 30, 40, 50]

max_value = reduce(lambda x, y: x if x > y else y, values)
print(max_value)
````

````python
# 5. Applying a Function to Each Element in a List
# Create a lambda function that squares each element in a list of numbers and apply it using the map() function.

numbers = [1, 2, 3, 4, 5]

squared_element = map(lambda x: x**2, numbers)

print(list(squared_element))
````

````python
# 6. Calculating Factorial of a Number
# Use a lambda function to calculate the factorial of a given number. Test it with the number 5.

# Meaning of 'factorial of a number:

'''
Cumulative multiplication of numbers from number to 1
'''
from functools import reduce

n = 5
result = reduce(lambda x, y: x * y, range(1, n+1))
print(result)
````
