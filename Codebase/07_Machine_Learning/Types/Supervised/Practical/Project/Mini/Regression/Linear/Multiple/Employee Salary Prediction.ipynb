### **2. Exercise: Multiple Linear Regression**
**Objective**: Predict the salary of an employee based on their years of experience and education level (years of schooling).

**Dataset**:  
| Experience (Years) | Education (Years) | Salary ($) |
|--------------------|-------------------|------------|
| 1                  | 12                | 40,000     |
| 2                  | 14                | 45,000     |
| 3                  | 16                | 55,000     |
| 5                  | 18                | 70,000     |
| 8                  | 20                | 90,000     |

**Steps**:
1. **Visualize the Data**: Plot Salary vs Experience and Salary vs Education using scatter plots.
2. **Apply Multiple Linear Regression**: Model the relationship using two predictor variables: Experience and Education.
   \[
   \text{Salary} = \beta_0 + \beta_1 (\text{Experience}) + \beta_2 (\text{Education})
   \]
3. **Prediction**: Predict the salary of an employee with 4 years of experience and 17 years of education.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
x = np.array([[1, 12], [2, 14], [3, 16], [5, 18], [8, 20]])  # Experience and Education
y = np.array([40000, 45000, 55000, 70000, 90000])  # Salary

# Create a multiple linear regression model
model = LinearRegression()

# Train the model
model.fit(x, y)

# Coefficients
print("Intercept (c):", model.intercept_)
print("Slope (m1, m2):", model.coef_)

# Predict for Experience=4 years, Education=17 years
predicted_salary = model.predict([[4, 17]])
print("Predicted Salary:", predicted_salary[0])

# Plotting Salary vs Experience
plt.figure(figsize=(12, 6))

# Subplot 1: Salary vs Experience
plt.subplot(1, 2, 1)
plt.scatter(x[:, 0], y, color='blue', label='Actual Data')  # Salary vs Experience
predicted_y = model.predict(x)  # Predicted Salary for all data points
plt.plot(x[:, 0], predicted_y, color='red', label='Regression Line (Experience)')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.grid(True)
plt.legend()

# Subplot 2: Salary vs Education
plt.subplot(1, 2, 2)
plt.scatter(x[:, 1], y, color='green', label='Actual Data')  # Salary vs Education
plt.plot(x[:, 1], predicted_y, color='orange', label='Regression Line (Education)')
plt.xlabel('Education (Years)')
plt.ylabel('Salary')
plt.title('Salary vs Education')
plt.grid(True)
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
