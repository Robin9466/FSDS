## **1. Linear Algebra**  

### **Definition**:  
Linear Algebra is the branch of mathematics that deals with vectors, matrices, and their transformations. It is foundational for handling multidimensional data and computations in machine learning and data science.

---

### **Topics Covered**:  

#### **Vectors, Matrices, and Matrix Operations**  
- **Meaning**:  
Vectors represent quantities with both magnitude and direction (e.g., `[1, 2, 3]`). Matrices are 2D arrays of numbers used to store and manipulate data efficiently. Operations like addition, multiplication, and transpose are essential for data manipulation.

- **Reasoning**:  
Used in data transformation, linear regression, and neural networks.  
For example: Converting features into mathematical models.  

- **Practical Application**:  
   - Representing datasets as matrices (rows = samples, columns = features).  
   - Transforming data with matrix multiplication (e.g., scaling features).  

#### **Eigenvalues and Eigenvectors**  
- **Meaning**:  
Eigenvalues measure the "stretch" of a transformation, while eigenvectors indicate the direction.  

- **Reasoning**:  
These are critical in reducing dimensionality (e.g., Principal Component Analysis).  

- **Practical Application**:  
   - Identifying important features in datasets.  
   - Reducing noise in large datasets for efficient computation.

- **Meaning**:  
Hypothesis testing determines if data supports a particular claim, and p-values measure the significance.  

- **Reasoning**:  
Used for validating models and experiments in data science.  

- **Practical Application**:  
   - Checking if a marketing campaign increases sales.  
   - Evaluating if two datasets come from the same distribution.

### **Guided Exercise: Gradient Descent**  

**Problem**: Write a Python function to perform a single step of gradient descent on a simple cost function, \( f(x) = x^2 \).  

**Solution (ChatGPT)**:  

```python
def gradient_descent(x, learning_rate):
    gradient = 2 * x  # Derivative of x^2 is 2x
    new_x = x - learning_rate * gradient
    return new_x

# Perform a step of gradient descent
x_initial = 10
learning_rate = 0.1
new_x = gradient_descent(x_initial, learning_rate)
print(f"Updated x after one step: {new_x}")
```

````python
import numpy as np

# Dataset as a matrix
scores = np.array([
    [85, 90, 78], # Student 1
    [88, 94, 80], # Student 2
    [84, 89, 76], # Student 3
])

# Calculate mean score per subject
mean_scores = np.mean(scores, axis=0)
print(f"Mean scores (Math, Science, English): {mean_scores}")
````

Your Exercise:

````python
history_column = np.array([95, 97, 78])

# Reshape the new column to a column vector
history_column = history_column.reshape(-1, 1)

# Append the new column
new_scores = np.append(scores, history_column, axis=1)

# Recalculate the mean scores for all subjects across all students
mean_scores1 = np.mean(new_scores, axis=0)
print(f"Mean scores (Math, Science, English, History): {mean_scores1}")
````

Mini-Project: PCA (Principal Component Analysis)
Guided Mini-Project:

Use a dataset with features like height, weight, and age.
Perform PCA to reduce the dataset to 2 dimensions.

Height,Weight,Age
170,70,25
160,65,30
180,75,35


````python
from sklearn.decomposition import PCA
import pandas as pd

# Create the dataset
data = pd.DataFrame({
      'Height': [170, 160, 180],
    'Weight': [70, 65, 75],
    'Age': [25, 30, 35]
})

# Perform PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

print("Reduced Data:\n", reduced_data)
````

Your Mini-Project:

Use a dataset with 4 features (e.g., Income, Expenses, Savings, Debt) and reduce it to 2 dimensions using PCA.
00	1800

````python
from sklearn.decomposition import PCA
import pandas as pd

# Create the dataset
data = {
    "Income": [5000, 7000, 8000, 6000, 7500, 9000, 10000, 11000, 9500, 10500],
    "Expenses": [2000, 2500, 3000, 2200, 2700, 3300, 4000, 4200, 3600, 3900],
    "Savings": [1500, 2000, 2500, 1800, 2100, 2800, 3500, 3800, 3000, 3200],
    "Debt": [1000, 1500, 1200, 800, 1300, 2000, 2500, 2200, 1600, 1800]
}

# Convert dictionary data to DataFrame
df = pd.DataFrame(data)

# Perform PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df)

print("Reduced Data:\n", reduced_data)

import matplotlib.pyplot as plt

plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.title("PCA Reduced Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
````

Generate a Poisson distribution with a mean of 5 and plot it.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Poisson distribution
mean = 5

# Generate Poisson distribution data
data = np.random.poisson(mean, 1000)

# Plot the distribution
plt.hist(data, bins=range(0, max(data)+2), density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title('Poisson Distribution (mean = 5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
