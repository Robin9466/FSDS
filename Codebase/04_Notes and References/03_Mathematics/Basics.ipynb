````python
def break_even_point(revenue_per_unit, cost_per_unit, fixed_cost):
    if revenue_per_unit <= cost_per_unit:
        return "Break-even point cannot be reached with these values"
    return fixed_cost / (revenue_per_unit - cost_per_unit)

# Example values
revenue_per_unit = 500
cost_per_unit = 200
fixed_cost = 1000

break_even_x = break_even_point(revenue_per_unit, cost_per_unit, fixed_cost)
print(f"Break-even point is at x = {break_even_x}")
````

Your Mini-Project:
Dataset:
Month, Sales
June, 2200
July, 2500
August, 2800
September, 3100
October, 3400

Task:
Calculate the average sales and percentage growth from June to October.
Visualize the sales data over time using Matplotlib.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sales Dataset
data = {
    'Month': ['June', 'July', 'August', 'September', 'October'],
    'Sales': [2200, 2500, 2800, 3100, 3400]
}

# Converting dataset from dictionary to DataFrame
df = pd.DataFrame(data)

# Calculate average sales and percentage growth from June to October
average_sales = df['Sales'].mean()
percentage_growth = ((df['Sales'].iloc[-1] - df['Sales'].iloc[0]) / df['Sales'].iloc[0]) * 100

print(f"Average sales: {average_sales}")
print(f"Percentage growth from June to October: {percentage_growth:.2f}%")

# Plotting sales data
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.title('Sales Growth Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()
````
