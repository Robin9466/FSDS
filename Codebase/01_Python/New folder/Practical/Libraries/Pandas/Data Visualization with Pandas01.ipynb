Exercise 1: Basic Plotting with Pandas
Data Preparation:

Create a DataFrame with columns Month (January to December) and Temperature (average monthly temperatures).
Plot a line chart showing the temperature trend over the year.
Objective: Practice basic plotting with Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data Preparation
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature': [5, 7, 10, 15, 20, 25, 30, 28, 24, 18, 10, 6]
}
df = pd.DataFrame(data)

# Plot Line Chart
df.plot(x='Month', y='Temperature', kind='line', marker='o')
plt.title('Monthly Temperature Trend')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()
```

Explanation:

The code creates a DataFrame with two columns: Month and Temperature.
df.plot() is used to generate the line chart. The x and y parameters specify the axes.
marker='o' adds markers to each data point. linestyle='--' creates a dashed line.
Customizations include title, labels, and grid for better readability.

Different Plot Types
Line Plot
Used for showing trends over time.

```python
# Another example: Plotting quarterly revenue
data = {
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Revenue': [1500, 1800, 2100, 2300]
}
df = pd.DataFrame(data)

# Plot
df.plot(x='Quarter', y='Revenue', kind='line', marker='D', color='blue')
plt.title('Quarterly Revenue')
plt.xlabel('Quarter')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.show()
```

Bar Chart
Used for comparing different categories.

```python
# Bar chart for daily sales in a week
data = {
    'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'Sales': [100, 120, 130, 140, 150, 160, 170]
}
df = pd.DataFrame(data)

# Plot
df.plot(x='Day', y='Sales', kind='bar', color='orange')
plt.title('Daily Sales')
plt.xlabel('Day')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

Histogram
Used for visualizing the distribution of data.

```python
import numpy as np

# Generating random age data
age_data = {'Age': np.random.randint(15, 75, 100)}
df = pd.DataFrame(age_data)

# Plot Histogram
df['Age'].plot(kind='hist', bins=10, color='purple', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Scatter Plot
Used for identifying relationships between variables.

```python
# Height vs Weight scatter plot
data = {
    'Height': [150, 160, 170, 180, 190, 200],
    'Weight': [50, 60, 70, 80, 90, 100]
}
df = pd.DataFrame(data)

# Plot Scatter Plot
df.plot(kind='scatter', x='Height', y='Weight', color='red', s=100)
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

Customizing Plots
Adding Titles, Labels, and Legends

```python
# Customizing the Temperature Line Chart
df.plot(x='Month', y='Temperature', kind='line', marker='o', color='darkcyan')
plt.title('Customized Temperature Trend')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)

# Adding legend
plt.legend(['Temperature Trend'], loc='upper left')
plt.show()
```
