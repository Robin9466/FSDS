```python
import pandas as pd

# Create a DataFrame
data = {
    'Name':['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 17, 22, 32],
    'Score': [85, 92, 88, 70]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Display the first 5 rows using head()
# Display the first 5 rows of the DataFrame
print(df.head())

# Display the last rows using tail().
# Display the last 5 rows of the DataFrame
print(df.tail())

# Get a summary of the DataFrame (data types, non-null counts, memory usage)
print(df.info())

# Get descriptive statistics of the DataFrame (like count, mean, std, min, max, etc.)
print(df.describe())

# Selecting specific columns and rows
# a) Selecting specific columns.
# Select specific columns (Name and Score)
selected_columns = df[['Name', 'Score']]
print(selected_columns)

# b) Selecting specific rows using iloc[] (integer-location based indexing).
# Select rows using iloc (rows 0 to 2)
selected_rows_iloc = df.iloc[0:3] # Note: Python is zero-indexed
print(selected_rows_iloc)

# Selecting specific rows and columns using loc[] (label-based indexing)
# Select rows with age > 25 and only display Name and Score columns
selected_rows_loc = df.loc[df['Age'] > 25, ['Name', 'Score']]
print(selected_rows_loc)
# Basic Operations
# Calculate mean of 'Age'
mean_age = df['Age'].mean()
print("Mean Age:", mean_age)

# Filter rows where Score is greater than 80
high_scores = df[df['Score'] > 80]
print("High Scorers:\n", high_scores)
```
