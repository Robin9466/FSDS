1. Basic Arithmetic Operations
NumPy supports element-wise arithmetic operations, which means you can perform calculations directly between arrays.

```python
# Creating sample arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Addition
addition = a + b

# Subtraction
subtraction = a - b

# Multiplication
multiplication = a * b

# Division
division = b / a

print("Addition between two array: ", addition)
print("Subtraction between the array a - b: ", subtraction)
print("Multiplication between two array: ", multiplication)
print("Division between array, 'a', and 'b': ", division)
```

2. Universal Functions (ufuncs)
Universal functions, or ufuncs, are built-in NumPy functions that operate element-wise on arrays.

```python
# Square root
sqrt = np.sqrt(a)

# Exponential
exp = np.exp(a)

# Logarithm (base e)
log = np.log(a)

# Sine, Cosine, etc.
sine = np.sin(a)

print(sqrt)
print(exp)
print(log)
print(sine)
```

3. Aggregate Functions
Aggregate functions perform calculations over the entire array, returning a single value (like the sum of all elements)

```python
# Aggregate Functions
# Sample array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Sum of all elements
total_sum = np.sum(arr)

# Mean of all elements
mean_value = np.mean(arr)

# Minimum and Maximum
min_value = np.min(arr)
max_value = np.max(arr)

# Standard Deviation
std_dev = np.std(arr)

print(total_sum)
print(mean_value)
print(min_value)
print(max_value)
print(std_dev)
```

4. Linear Algebra Operations
NumPy is excellent for matrix operations, making it a core library for scientific computing.

```python
# Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Dot Product
dot_product = np.dot(matrix1, matrix2)

# Transpose of a matrix
transpose = np.transpose(matrix1)

# Determinant of a matrix
det = np.linalg.det(matrix1)

# Inverse of a matrix
inverse = np.linalg.inv(matrix1)

print(dot_product)
print(transpose)
print(det)
print(inverse)
```

Array Manipulation
Array manipulation in NumPy allows you to reshape, merge, and break down arrays, enabling flexibility in handling numerical data.

Stacking & Splitting
Stacking involves combining multiple arrays into one:

```python
import numpy as np

# Example arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Vertrically stack 'a' and 'b'
result = np.vstack((a, b))
print(result)
```

```python
# Horizontally stack 'a' and 'b'
result = np.hstack((a, b))
print(result)
```

```python
# Concatenate 'a' and 'b' along the default axis (0)
result = np.concatenate((a, b))
print(result)
```

Splitting divides an array into multiple sub-arrays:

```python
c = np.array([1, 2, 3, 4, 5, 6])

# Split 'c' into 3 equal parts
result = np.split(c, 3)
print(result)
```

array_split: Allows splitting into unequal parts.

```python
# Split 'c' into 4 parts
result = np.array_split(c, 4)
print(result)
```

Mini-Project: Create a data merger that stacks student scores (row-wise) from two different tests and then splits them by groups of students.

```python
import numpy as np

# Step 1: Creating mock data using Numpy arrays
# Data for Test 1 (5 students, 2 subjects)
data_test1 = np.array([
    [85, 90],  # S1 - [Math Score, English Score]
    [78, 82],  # S2
    [92, 89],  # S3
    [88, 85],  # S4
    [76, 80]   # S5
])

data_test2 = np.array([
    [91, 88],  # S6
    [74, 79],  # S7
    [89, 92],  # S8
    [95, 87],  # S9
    [80, 81]   # S10
])

# Step 2: Merge Data
merged_data = np.vstack((data_test1, data_test2))

# Step 3: Splitting
group_size = 5
student_groups = np.array_split(merged_data, len(merged_data)//group_size)
print(type(student_groups))
# Display each group using enumerate
for index, group in enumerate(student_groups):
    print(f"Group {index + 1}:")
    print(group)
    print("\n")
```

Fancy Indexing & Boolean Masking
Fancy indexing lets you select specific elements using arrays of indices, while boolean masking allows selection based on conditions.

```python
# Fancy Indexing
# Using arrays to index other arrays:
data = np.array([10, 20, 30, 40, 50])
indices = [0, 2, 4]

# Select elements at indices 0, 2, and 4
result = data[indices]
print(result)
```

Boolean Masking:
Selecting elements based on conditions:

```python
data = np.array([10, 20, 30, 40, 50])

# Create a boolean mask for values greater than 25
mask = data > 25
result = data[mask]
print(result)
```

Mini-Project: Given a list of students and their scores, use boolean masking to filter out students who passed a test (score >= 60) and highlight the top scorers using fancy indexing.



```python
import numpy as np

# Step 1: Create lists of student names and their corresponding scores
students = np.array(["Alice", "Bob", "Charlie", "David", "Eve"])
scores = np.array([58, 92, 75, 85, 60])

# Step 2: Use boolean masking to filter students who passed (score >= 60)
passed_mask = scores >= 60
passed_students = students[passed_mask]
passed_scores = scores[passed_mask]

print("Students who passed:")
print(passed_students)
print(passed_scores)

# Step 3: Highlight top scorers (e.g., top 3 students)
# We can sort the scores in descending order and then use fancy indexing to select the top scorers
top_scorer_indices = np.argsort(scores)[-3:]  # Top 3 scorers
top_students = students[top_scorer_indices]
top_scores = scores[top_scorer_indices]

print("\nTop scorers:")
print(top_students)
print(top_scores)
```

Statistical Operations on Arrays
Statistical operations help you extract insights from datasets.

Key Statistical Methods
Mean (np.mean): Average value.

Median (np.median): Middle value of sorted data.

Standard Deviation (np.std): Spread of data around the mean.

Variance (np.var): Square of standard deviation.

```python
data = np.array([1, 2, 3, 4, 5])

# Calculate statistics
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
variance = np.var(data)

# Displaying the values
print(mean)
print(median)
print(std_dev)
print(variance)
```

```python
'''
Exercise:

Generate a dataset of random numbers and compute mean, median, standard deviation, and variance. Create a histogram using numpy.histogram
'''
import numpy as np
import matplotlib.pyplot as plt
# Step 1: Generate dataset of random integers between 1 and 100, size of 100
data = np.random.randint(1, 10, size = 100)
print(data)

# Step 2: Calculating statistical measures
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
variance = np.var(data)

# Printing statistical results
print(mean)
print(median)
print(std_dev)
print(variance)

# Step 3: Create a histogram using numpy.histogram
hist, bin_edges = np.histogram(data, bins=10) # Using 10 bins for the histogram

# Display histogram data
print("Histogram counts:", hist)
print("Bin edges:", bin_edges)

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=10, edgecolor='black', alpha=0.7, color='green')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
