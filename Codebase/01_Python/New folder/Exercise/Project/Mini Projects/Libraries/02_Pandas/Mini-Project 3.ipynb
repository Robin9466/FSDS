Mini-Project 3:
Use a sales dataset (e.g., "Product", "Region", "Sales Amount").
Group by "Region" and calculate the total sales.
Create a pivot table to see sales per "Product" in each "Region".

Step-by-Step Plan

1. Load the Data:

Import the dataset into a DataFrame using pandas.
Inspect the dataset to understand its structure and ensure it has the required columns.

2. Group by Region:

Use the groupby function to calculate the total sales amount per region.

3. Create Pivot Table:

Utilize pandas' pivot table functionality to create a table with:
Region as rows.
Product as columns.
The sum of Sales Amount as the values.

```python
import pandas as pd

# Load the data
df = pd.read_csv('sales_data.csv')
df
```

```python
# Group by Region to calculate total sales
total_sales_by_region = df.groupby('Region')['Sales Amount'].sum().reset_index()

# Display total sales by region
print("Total Sales by Region:")
total_sales_by_region
```

```python
# Create a Pivot Table to see sales per Product in each Region
pivot_table = pd.pivot_table(df, values='Sales Amount', index='Region', columns='Product', aggfunc='sum', fill_value=0)

# Display the Pivot Table
print(df)
print()
print("\nPivot Table - Sales per Product in each Region")
pivot_table
```
