### **Assignments**

#### **Module Assignment:**
1. Create a Python file named `math_operations.py`.
2. Inside `math_operations.py`, define functions for basic arithmetic operations: `add`, `subtract`, `multiply`, and `divide`.
3. In a separate file, import your `math_operations` module and use its functions to perform some calculations.

#### **Exception Handling Assignment:**
1. Write a Python script that asks the user for two numbers and performs division.
2. Handle the following exceptions:
   - `ValueError` if the user enters non-numeric input.
   - `ZeroDivisionError` if the user tries to divide by zero.
3. Ensure that appropriate messages are displayed for each type of exception.
