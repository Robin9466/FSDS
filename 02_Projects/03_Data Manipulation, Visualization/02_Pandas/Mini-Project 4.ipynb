Mini-Project: Visualizing Daily Traffic Data

Problem:
You are working as a data analyst at a traffic monitoring company. Your task is to create a comprehensive visualization of daily traffic data for a week. You have the number of vehicles that passed through a specific point on the highway for each day.

Data:

Days: 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
Vehicles: [1500, 2000, 1800, 2200, 2500, 2700, 2300]
Instructions:

Line Plot:
Plot a line chart showing the number of vehicles for each day.
Add markers, a title, and axis labels.
Bar Chart:
Create a bar chart showing the traffic for each day of the week.
Customize the color of the bars and rotate the x-axis labels for readability.
Histogram:
Create a histogram to show the distribution of traffic values (i.e., the frequency of certain traffic ranges).
Scatter Plot:
Create a scatter plot comparing the number of vehicles on weekdays (Monday to Friday) with the weekend (Saturday and Sunday).
Customization:
Add gridlines, titles, and adjust the style for better aesthetics.

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Set up data

data = {
    'Days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'Vehicle': [1500, 2000, 1800, 2200, 2500, 2700, 2300]
}

# Converting dictionary to dataframe
df = pd.DataFrame(data)
df
```

Line Plot:
Plot a line chart showing the number of vehicles for each day.
Add markers, a title, and axis labels.

```python
# Plotting Line Plot
df.plot(x='Days', y='Vehicle', kind='line', marker='o')
plt.title('Days vs Vehicle')
plt.xlabel('Days')
plt.ylabel('Vehicles')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
```

Bar Chart:
Create a bar chart showing the traffic for each day of the week.
Customize the color of the bars and rotate the x-axis labels for readability.

```python
# Bar Chart
df.plot(x='Days', y='Vehicle', kind='bar', color='orange', edgecolor='black')
plt.title('Days vs Vehicle')
plt.xlabel('Days')
plt.ylabel('Vehicle')
plt.grid(True)
plt.xticks(rotation=45)
```

Histogram:
Create a histogram to show the distribution of traffic values (i.e., the frequency of certain traffic ranges).


```python
# Data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
Vehicle = [1500, 2000, 1800, 2200, 2500, 2700, 2300]

# Plotting histogram
plt.figure(figsize=(8, 5))
plt.hist(Vehicle, bins= 5, color='skyblue', edgecolor='black', alpha=0.7)

# Add lables and title
plt.title('Traffic Distribution Histogram', fontsize=14)
plt.xlabel('Traffic Range (Number of Vehicles)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

Scatter Plot:
Create a scatter plot comparing the number of vehicles on weekdays (Monday to Friday) with the weekend (Saturday and Sunday).

```python
import matplotlib.pyplot as plt

# Data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
vehicles = [1500, 2000, 1800, 2200, 2500, 2700, 2300]

# Splitting data into weekdays and weekends
weekdays = days[:5]
weekend = days[5:]
weekdays_vehicles = vehicles[:5]
weekend_vehicles = vehicles[5:]

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(weekdays, weekdays_vehicles, color='green', label='Weekdays', s=100, edgecolor='black', marker='o')
plt.scatter(weekend, weekend_vehicles, color='red', label='Weekend', s=100, edgecolor='black', marker='o')

# Adding details to the plot
plt.title('Number of Vehicles: Weekdays vs Weekend', fontsize=16)
plt.xlabel('Days of the Week', fontsize=14, color='blue')
plt.ylabel('Number of Vehicles', fontsize=14, color='orange')
plt.legend(loc='upper left')

# Adding gridlines for better visibility
plt.grid(True, linestyle='--', alpha=0.6, which='both')

# Adjusting plot style
plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```
