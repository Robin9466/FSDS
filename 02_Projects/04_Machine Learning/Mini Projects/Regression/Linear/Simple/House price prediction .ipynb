### **1. Exercise: Simple Linear Regression**
**Objective**: To predict the price of a house based on its size (square footage).

**Dataset**:  
| House Size (sq ft) | House Price (in $) |
|--------------------|--------------------|
| 1500               | 300,000            |
| 1800               | 350,000            |
| 2400               | 500,000            |
| 3000               | 600,000            |
| 3500               | 700,000            |

**Steps**:
1. **Visualize the Data**: Plot a scatter plot of House Size vs House Price.
2. **Apply Linear Regression**: Use simple linear regression to find the line of best fit. The equation will be in the form:
   \[
   \text{House Price} = \beta_0 + \beta_1 (\text{House Size})
   \]
3. **Prediction**: Use the model to predict the price of a house with 2500 sq ft.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
x = np.array([1500, 1800, 2400, 3000, 3500]).reshape(-1, 1) # House Size
y = np.array([300000, 350000, 500000, 600000, 700000]) # House price

# Create a Linear regression model
model = LinearRegression()

# Train the model
model.fit(x, y)

# Coefficients
print("Intercept (β0):", model.intercept_)
print("Slope (β1):", model.coef_)

# Predict for 2500 sq ft
predicted_price = model.predict([[2500]])
print("Predicted Price for 2500 sq ft:", predicted_price[0])

# Plotting
plt.scatter(x, y, color='blue')
plt.grid(True)
plt.plot(x, model.predict(x), color='red')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($)')
plt.title('House Size vs House Price')
plt.show()
```
