```python
x = df['TV']
x
```

```python
y = df['Sales']
y
```

```python
from sklearn.linear_model import LinearRegression # Changed 'liner_model' to 'linear_model'
```

```python
# Create a LinearRegression object
linear = LinearRegression()

# Fit the model using your data
# Reshape x to a 2D array using reshape(-1, 1)
linear.fit(x.values.reshape(-1, 1), y)
```

```python
linear.intercept_
```

```python
linear.coef_
```

```python
file = 'linear_reg.sav'
# Open the file in write binary ('wb') mode using 'with open(...)'
# This ensures the file is automatically closed even if errors occur.
with open(file, 'wb') as f:
    pickle.dump(linear, f)
```

```python
linear.predict([[45]])
```
