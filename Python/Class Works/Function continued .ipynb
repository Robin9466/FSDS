```python
#     q1 : Try to print a prime number in between 1 to 1000

def is_prime(n):
  if n <=1:
    return False
  if n <=3:
    return True

  for i in range(2, int(n**0.5)+1):
    if n % i == 0:
      return False
  return True


prime_numbers = [num for num in range(1,1001) if is_prime(num)]
print("Prime numbers from 1 to 1000 =", end = " ")
print(prime_numbers)
```

```python
# Replica of list append

def replica_append(my_list, arg):
    my_list[len(my_list):] = [arg]
    return my_list

# Define a list
my_list = [1, 2, 3]

# Call the replica_append function to add an element
new_list = replica_append(my_list, 4)

# Print the new list
print(new_list)
```

```python
# Replica of extend function

def replica_extend(my_list, *args):
    for arg in args:
        my_list.append(arg)
    return my_list
```

```python
# Replica of pop function

def replica_pop(my_list, index=None):
    if index is None:
        # Handle the case where index is not provided
        if my_list:
            del my_list[-1]  # Remove the last element
    elif 0 <= index < len(my_list):
        # Check if the index is valid
        del my_list[index]
    return my_list
```

```python
# 1. **Sort List of Tuples:** Sort a list of tuples based on the second element of each tuple using a lambda function.

list_of_tuples = [(3,8), (7,8), (4,7),(True, False)]

# Sorting order would be ascending:

key_function = lambda x : x[1]
sorted_tuples_asc = sorted(list_of_tuples, key = key_function)

# Sorting order would be descending:
sorted_tuples_des = sorted(list_of_tuples, key = key_function, reverse = True)

print(f"List of tuples in ascending and descending order respectively : {sorted_tuples_asc},{sorted_tuples_des}")
```

```python
# 4. **Filter Palindromes:** Filter out palindromic words from a list of strings using a lambda function.

my_list = ["mim", "Mom", "Robin", "radar", "tat", "tit"]

is_palindrome = lambda word : word.lower() == word.lower()[::-1]
filtered_palindrome = list(filter(is_palindrome, my_list))
print(f" Filtered palindromes are : {filtered_palindrome}")
```

```python
# 7. **Find Prime Numbers:** Use a lambda function to generate a list of prime numbers within a given range.

# To check whether number is prime or not.
is_prime = lambda num : num > 1 and all(num%i != 0 for i in range(2, int(num**0.5)+1))

# Taking inputs from the user for start and end value

input_str = input(f"Provide the value of start and end respectively to find prime numbers in between, Seperated by space")
start_str, end_str = input_str.split()
start = int(start_str)
end = int(end_str)

# Use a list comprehension with the lambda function to find prime numbers within the range
prime_numbers = [num for num in range(start, end+1) if is_prime(num)]

# Print the prime numbers
print(f"Prime numbers from {start} to {end} are {prime_numbers}")
```

```python
# 8.  **Calculate BMI:** Calculate the Body Mass Index (BMI) of a person using a lambda function.

# Taking inputs of height, and weight from the user

height = float(input(f"Enter height in metre : "))
weight = float(input(f"Enter weight in kilogram : "))

# Declare lambda function to calculate BMI

calculate_BMI = lambda height, weight : weight/height**2

# Call function to calculate the BMI

BMI = calculate_BMI(height, weight)

# Print the result(BMI)
print(f"your BMI is {BMI:.2f}")

"""
Underweight: BMI less than 18.5
Normal weight: BMI between 18.5 and 24.9
Overweight: BMI between 25 and 29.9
Obesity (Class I): BMI between 30 and 34.9
Obesity (Class II): BMI between 35 and 39.9
Obesity (Class III): BMI 40 or greater
"""

# Classification of the BMI

if BMI < 18.5:
    print(f"Oops, You are underweight")
elif 18.5 <= BMI <= 24.9:
    print(f"Congratulations! Your weight is normal")
elif 25.0 <= BMI <= 29.9:
    print(f"Oh! You need to care, You are overweight")
elif 30.0 <= BMI <= 34.9:
    print(f"OMG, You need to be very careful, As you are having Obesity with Class 1")
elif 35.0 <= BMI <= 39.9:
    print(f"OMG, You need to be very careful, As you are having Obesity with Class 2")
elif BMI >= 40:
    print(f"OMG, You need to be very careful, As you are having Obesity with Class 3")
```

```python
import string
# 9. **Reverse Strings:** Reverse a list of strings using a lambda function.

# Take a list of strings from the user

number_of_strings = int(input(f"How many strings do you want to enter in a list? "))

# Use list comprehension to take the input from the user

list_of_strings = [input(f"Enter a string ") for _ in range(number_of_strings)]

# Use the lambda function to reverse a list of strings

reversed_list_strings = lambda list_of_strings : list_of_strings[::-1]

# Call the function

result = reversed_list_strings(list_of_strings)

# Print the original and the reversed list
print(f"Original list of strings are {list_of_strings}")
print(f"Reveresed list of strings are {result}")
```

```python
# 10. **Find Square Root:** Compute the square root of a number using a lambda function.

# Import module for square root

from math import sqrt

# Take a number from the user with input validation

while True:
    try:
        number = float(input(f"Enter a non negative number "))
        if number < 0:
            print(f"Kindly input a non-negative number")
        else:
            break # Exit the loop if a valid number is entered
    except ValueError:
        print(f"Invalid Input, Kindly input valid number")

# Use a lambda function to find the square root of a number

square_root = lambda number : sqrt(number)

# Store the value of the square root of the given number

result = square_root(number)

# Print the result

print(f"The square root of the {number} is {result:.2f}")
```

```python
# a. **Check Leap Year:** Determine whether a given year is a leap year or not using a lambda function.

# Take user-input of a year

year = int(input(f"Enter the year : "))

# Use lambda function to check for the leap year

leap_year = lambda year : (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

# Check for the leap year and provide user-friendly output

if leap_year(year):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```python
# e. **Calculate Area of Shapes:** Calculate the area of various shapes (circle, rectangle, triangle) using a lambda function.
# Circle
radius = float(input("Enter the radius of the circle: "))
area_of_circle = lambda r: 3.14 * r ** 2
print(f"Area of the circle is {area_of_circle(radius):.2f}")

# Rectangle
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
area_of_rectangle = lambda l, w: l * w
print(f"Area of the rectangle is {area_of_rectangle(length, width):.2f}")

# Triangle
base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))
area_of_triangle = lambda b, h: 0.5 * b * h
print(f"Area of the triangle is {area_of_triangle(base, height):.2f}")
```

```python
# f. **Sort Dictionaries:** Sort a list of dictionaries based on a specific key using a lambda function.

# Initialize an empty list to store dictionaries
list_of_dicts = []

# Get the number of dictionaries from the user
num_dicts = int(input("Enter the number of dictionaries: "))

# Use a list comprehension to take input for each dictionary
list_of_dicts = [
    {
        "name": input("Enter the name: "),
        "price": float(input("Enter the price: ")),
        "quantity": int(input("Enter the quantity: "))
    }
    for _ in range(num_dicts)
]

# Print the list of dictionaries
print("List of dictionaries:")
for item in list_of_dicts:
    print(item)

# Sort the list of dictionaries based on the "price" attribute
sorted_list = sorted(list_of_dicts, key=lambda x: x["price"])

# Display the sorted list
print(f"Sorted list based on price is: {sorted_list}")
```
