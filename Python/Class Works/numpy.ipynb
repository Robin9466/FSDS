```python
df2 = sns.load_dataset('titanic')
```

```python
df2.head()
```

```python
import plotly.graph_objects as go
import plotly.io as pio

# Assuming df2 is your DataFrame
fig = go.Figure(data=go.Scatter(x=df2['age'], y=df2['fare'], mode='markers'))

# Set layout
fig.update_layout(
    title='Scatter Plot of Age vs Fare',
    xaxis=dict(title='Age'),
    yaxis=dict(title='Fare'),
)

# Use plotly.io to display the plot in Colab
pio.show(fig)
```

```python
import plotly.express as px

# Assuming df2 is your DataFrame
fig = px.scatter(df2, x='age', y='fare', title='Scatter Plot of Age vs Fare')
fig.show()
```
