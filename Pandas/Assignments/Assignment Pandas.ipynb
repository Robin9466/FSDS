```python
# 1 . Find out their average rating on a weekly basis keep this in mind that they take two days of leave

import pandas as pd

# Loading the dataset from the CSV file
df = pd.read_csv(r'AgentPerformance.csv')

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group by 'Agent Name' and 'Date', then calculate the mean for 'Average Rating'
weekly_average_rating = df.groupby(['Agent Name', pd.Grouper(key='Date', freq='W-Mon', closed='right', label='right')])['Average Rating'].mean().reset_index()

# Print or use the resulting DataFrame
print(weekly_average_rating)
```

```python
# 3. Total query that you have taken
import pandas as pd

# Loading the dataset from the CSV file
df = pd.read_csv(r'AgentPerformance.csv')

# Group by 'Agent Name' and calculate the sum for 'Total Chats'
total_chats_per_agent = df.groupby('Agent Name')['Total Chats'].sum().reset_index()

# Add 'SL No' column using cumulative sum
total_chats_per_agent['SL No'] = total_chats_per_agent.groupby('Agent Name')['Total Chats'].cumsum() - total_chats_per_agent['Total Chats'] + 1

# Reorder columns
total_chats_per_agent = total_chats_per_agent[['SL No', 'Agent Name', 'Total Chats']]

# Print or use the resulting DataFrame
print(total_chats_per_agent)
```

```python
# 4. Total feedback that you have received

total_feedback = df.groupby('Agent Name')['Total Feedback'].sum()
print(total_feedback)
```

```python
# 9. Average weekly response time for each agent
# Average weekly response time for each agent
import pandas as pd

# Assuming 'Date' is a datetime column in your DataFrame
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index (inplace=True is important here)
df.set_index('Date', inplace=True)

# Convert 'Average Response Time' to timedelta
df['Average Response Time'] = pd.to_timedelta(df['Average Response Time'])

# Calculate the average weekly response time for each agent
result = df.groupby('Agent Name')['Average Response Time'].resample('W').mean().reset_index()

# Reset the index to move 'Week' from the index to a regular column
result.reset_index(drop=True, inplace=True)

# Convert timedelta to hr:mn:sec format
result['Average Response Time'] = result['Average Response Time'].apply(lambda x: str(x).split()[-1])

# Select rows for unique agents
unique_agents_result = result[~result['Agent Name'].duplicated(keep='first')]

# Print the result or perform any additional actions
unique_agents_result[['Agent Name', 'Average Response Time']]
```

```python
# 10. average weekly resolution time for each agent 
# Assuming df is your DataFrame
# Convert 'Average Resolution Time' to timedelta if it's not already
df['Average Resolution Time'] = pd.to_timedelta(df['Average Resolution Time'])

# Group by 'Agent Name' and 'Date' and calculate the average weekly resolution time
Average_Resolution_Time_Weekly = df.groupby(['Agent Name', df.index.to_period('W-Mon')])['Average Resolution Time'].mean()

# Reset the index to make 'Agent Name' and 'Date' regular columns
result_df = Average_Resolution_Time_Weekly.reset_index()

# Group by 'Agent Name' and calculate the overall average weekly resolution time for each agent
average_resolution_time_unique_agents = result_df.groupby('Agent Name')['Average Resolution Time'].mean()

# Create a new DataFrame with unique agents and their average weekly resolution time
average_resolution_time_unique_agents_df = pd.DataFrame(average_resolution_time_unique_agents).reset_index()
average_resolution_time_unique_agents_df.columns = ['Agent Name', 'Average_Resolution_Time_Weekly']

# Print the DataFrame
print(average_resolution_time_unique_agents_df)
```

```python
# 11. list of all agents' names 

unique_agents = pd.DataFrame(df['Agent Name'].unique(), columns=['Agent Name'])
print(unique_agents)
```

```python
# 12. percentage of chats on which they have received feedback 
percentage_chats_feedback = df['Total Feedback'] / df['Total Chats'] * 100
df['percentage_chats_feedback'] = percentage_chats_feedback
result_df = df[['Agent Name', 'percentage_chats_feedback']].copy()
print(result_df)
```

```python
# 13 . Total contribution hour for each and every agent's weekly basis

import pandas as pd

# Assuming df is your DataFrame

# Convert 'Average Resolution Time' to timedelta if it's not already
df['Average Resolution Time'] = pd.to_timedelta(df['Average Resolution Time'])

# Group by 'Agent Name' and calculate the total contribution hours for each week
contribution_hour = df.groupby('Agent Name')['Average Resolution Time'].resample('W').sum()

# Reset the index to make 'Agent Name' and 'Date' regular columns
result_df = contribution_hour.reset_index()

# Rename the columns
result_df.columns = ['Agent Name', 'Date', 'Contribution Hours']  # Adjusted column name

print(result_df)

```

```python
# 14. total percentage of active hours for a month 
import pandas as pd

# Assuming df is your DataFrame
# Convert 'Average Resolution Time' to timedelta if it's not already
df['Average Resolution Time'] = pd.to_timedelta(df['Average Resolution Time'])

# Group by 'Agent Name' and calculate the total contribution hours for each week
contribution_hour = df.groupby(['Agent Name', df.index.to_period('W-Mon')])['Average Resolution Time'].sum()

# Reset the index to make 'Agent Name' and 'Date' regular columns
result_df = contribution_hour.reset_index()

# Calculate the total percentage of active hours for each agent in a month
total_active_hours_month = result_df.groupby('Agent Name')['Average Resolution Time'].sum()

# Create a new DataFrame with unique agents and their total contribution hours for the month
total_active_hours_month_df = pd.DataFrame(total_active_hours_month).reset_index()
total_active_hours_month_df.columns = ['Agent Name', 'Total Active Hours in Month']

# Print the DataFrame
print(total_active_hours_month_df)
```
