```python
# map, reduce, filter
```

```python
l = [2,3,4,5,6]
```

```python
l1 = []
for i in l:
    l1.append(i*2)
```

```python
l1
```

```python
def sq(n):
    return n*2
```

```python
list(map(sq,l))
```

```python
list(map(lambda x : x*2, l))
```

```python
l
```

```python
list(map(lambda x : str(x),l))
```

```python
def str_c(n):
    return str(n)
```

```python
list(map(str_c,l))
```

```python
list(map(str,l))
```

```python
from functools import reduce
```

```python
l = [3]
```

```python
sum(l)
```

```python
reduce(lambda a,b: a*b,l)
```

```python
l = [1,2,3,3,45,5]
```

```python
def check_even(n):
    if n%2 == 0:
        return True
```

```python
list(filter(check_even,l))
```
